```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#url for web scraping 

url = "https://timesofindia.indiatimes.com/entertainment/tamil/movie-reviews?curpg=3"

# access the url with request object

result = requests.get(url)

#scraping webpage content
webpage = result.content

#create a soup object for parsing web page using html parser
sl_soup = BeautifulSoup(webpage, 'html.parser')

```

```python
# close the result object
result.close()
```

```python
sl_soup.contents
```

```python
entire_table = sl_soup.findAll('div', attrs = {'class':'FIL_right'})
```

```python
entire_table[0]
```

```python
# extracting movie names from the entire page

movie_title = []
for movie_name in entire_table:
    #print (movie_name.h3.text)
    name = movie_name.h3.text
    #print(name)
    movie_title.append(name)
```
